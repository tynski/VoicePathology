```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Parameters Vector

Load, merge, normalization vector of parameters.

```python
# Loading each csv file
df1 = pd.read_csv('fh.csv')
df2 = pd.read_csv('fp.csv')
```

Concatenating all four DataFrames into one.

```python
allpatients = pd.concat([df1,df2])
allpatients.head()
```

Features has diffrent rows, so they have to be normalized.

```python
features = ['MEAN','MAX','MIN','RMS','ENERGY']
for i in features:
    allpatients[i] = (allpatients[i] - np.mean(allpatients[i]))/np.std(allpatients[i])
allpatients.head()
```

```python
sns.pairplot(allpatients, hue='Status', palette='bwr')
```

```python
bins = 50
fig = plt.figure()
fig.set_size_inches(16, 10.5)
plt.subplot(3,2,1)
plt.hist(allpatients[allpatients['Status']=='healthy']['MAX'], bins, alpha=0.5, label='zdrowy')
plt.hist(allpatients[allpatients['Status']=='pathology']['MAX'], bins, alpha=0.5, label='chory')
plt.legend(loc='upper right')
plt.xlabel('Maksimum')
plt.ylabel('Liczno')
plt.subplot(3,2,2)
plt.hist(allpatients[allpatients['Status']=='healthy']['MIN'], bins, alpha=0.5, label='zdrowy')
plt.hist(allpatients[allpatients['Status']=='pathology']['MIN'], bins, alpha=0.5, label='chory')
plt.legend(loc='upper right')
plt.xlabel('Minimum')
plt.ylabel('Liczno')
plt.subplot(3,2,3)
plt.hist(allpatients[allpatients['Status']=='healthy']['RMS'], bins, alpha=0.5, label='zdrowy')
plt.hist(allpatients[allpatients['Status']=='pathology']['RMS'], bins, alpha=0.5, label='chory')
plt.legend(loc='upper right')
plt.xlabel('RMS')
plt.ylabel('Liczno')
plt.subplot(3,2,4)
plt.hist(allpatients[allpatients['Status']=='healthy']['ENERGY'], bins, alpha=0.5, label='zdrowy')
plt.hist(allpatients[allpatients['Status']=='pathology']['ENERGY'], bins, alpha=0.5, label='chory')
plt.legend(loc='upper right')
plt.xlabel('Energia')
plt.ylabel('Liczno')
plt.subplot(3,2,5)
plt.hist(allpatients[allpatients['Status']=='healthy']['MEAN'], bins, alpha=0.5, label='zdrowy')
plt.hist(allpatients[allpatients['Status']=='pathology']['MEAN'], bins, alpha=0.5, label='chory')
plt.legend(loc='upper right')
plt.xlabel('rednia')
plt.ylabel('Liczno')
```

Afterwards data have to be suffled, in order to avoid overfitting

```python
allpatients = allpatients.sample(frac=1).reset_index(drop=True)
allpatients.head()
```

Exploratory Data Analysis, quick look at value of each parameters.

We'll need to convert categorical features (Status) to dummy variables, because our alghoritm can't understant what means 'healthy'.

```python
status = pd.get_dummies(allpatients['Status'],drop_first=True)
```

```python
allpatients.drop(['Status'],axis=1,inplace=True)
```

```python
allpatients['Pathology'] = status
allpatients.head()
```

# Detection alghoritm

Building logistic regression algorithm

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(allpatients.drop(['Pathology'],axis=1), 
                                                    allpatients['Pathology'], 
                                                    test_size=0.20, 
                                                    random_state=42)
```

# Training and predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression(max_iter=100, C = 0.001)
logModel.fit(X_train, y_train)
```

```python
prediction = logModel.predict(X_test)
```

# Evaluation
Check precission, recall, f1-score using classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```
