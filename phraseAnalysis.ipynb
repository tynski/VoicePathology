```python
from scipy.io import wavfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sampleRate = wavfile.read('./voiceDataset/2-phrase-egg.wav')[0]
sampleTime = 1/sampleRate
n = len(wavfile.read('./voiceDataset/2-phrase-egg.wav')[1])
time = np.arange(0,n)*sampleTime
```

```python
phraseDict = {'phraseEGG' : wavfile.read('./voiceDataset/2-phrase-egg.wav')[1],
              'phrase' : wavfile.read('./voiceDataset/2-phrase.wav')[1]
             }
```

```python
phraseDf = pd.DataFrame(data = phraseDict, index = time)
```

```python
fig, axes = plt.subplots(2,1)
fig.set_size_inches(11.7, 8.27)
sns.lineplot(data = phraseDf['phraseEGG'], ax = axes[0]).set_title('phraseEGG')
sns.lineplot(data = phraseDf['phrase'], ax = axes[1]).set_title('phrase')
```
