```python
from scipy.io import wavfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sampleRate = wavfile.read('./voiceDataset/2-a_n-egg.wav')[0]
sampleTime = 1/sampleRate
n = len(wavfile.read('./voiceDataset/2-a_n-egg.wav')[1])
time = np.arange(0,n)*sampleTime
```

```python
vowelsDict = {'aVowelEGG' : wavfile.read('./voiceDataset/2-a_n-egg.wav')[1],
              'aVowel' : wavfile.read('./voiceDataset/2-a_n.wav')[1]
             }
```

```python
vowelsDf = pd.DataFrame(data = vowelsDict, index = time)
```

```python
fig, axes = plt.subplots(2,1)
fig.set_size_inches(11.7, 8.27)
#sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 0.5})
sns.lineplot(data = vowelsDf['aVowelEGG'], ax = axes[0]).set_title('aVowelEGG')
sns.lineplot(data = vowelsDf['aVowel'], ax = axes[1]).set_title('aVowel')
```
