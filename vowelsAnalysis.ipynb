```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
aLowFemale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/3-a_l.wav')[1]
aNaturalFemale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/3-a_n.wav')[1]
aHighFemale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/3-a_h.wav')[1]
aLowMale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/4-a_l.wav')[1]
aNaturalMale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/4-a_n.wav')[1]
aHighMale = wavfile.read('/home/barti/Documents/voicePathology/voiceDataset/4-a_h.wav')[1]
```

```python
fig, axes = plt.subplots(3, 1)
fig.set_size_inches(14, 9)
sns.lineplot(data = aLowFemale, ax = axes[0]).set_title('aLowFemale')
sns.lineplot(data = aNaturalFemale, ax = axes[1]).set_title('aNaturalFemale')
sns.lineplot(data = aHighFemale, ax = axes[2]).set_title('aHighFemale')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(11.7, 8.27))
sns.lineplot(ax = axes[0], data = aLowMale).set_title('aLowMale')
sns.lineplot(ax = axes[1], data = aNaturalMale).set_title('aNaturalMale')
sns.lineplot(ax = axes[2], data = aHighMale).set_title('aHighMale')
```
